```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from time import sleep
import csv
```

```python
s = Service("../chromedriver/chromedriver.exe")
driver = webdriver.Chrome(service=s)
driver.get("https://worldathletics.org/competition/calendar-results?hideCompetitionsWithNoResults=true&disciplineId=2&query=marathon&rankingCategoryId=5")
results = driver.find_element(By.CLASS_NAME, "ResultsTable_link__2ZKlr")

table = driver.find_element(By.CLASS_NAME, "ResultsTable_resultsTable__JBH1Y")
body = table.find_element(By.TAG_NAME, "tbody")
rows = body.find_elements(By.TAG_NAME, "tr")
# for row in rows:
#     link = row.find_element(By.TAG_NAME, 'a')
#     print(link)
sleep(5)
# marathon_name = rows[0].find_element(By.TAG_NAME, "span").text
# print(marathon_name)
# link = rows[0].find_element(By.TAG_NAME, "a")
# print(link)
# link.click()

data = []
for row in rows:
    marathon_name = row.find_element(By.TAG_NAME, "span").text
    date = row.find_element(By.TAG_NAME, "td").text
    link = row.find_element(By.PARTIAL_LINK_TEXT, "Result")
    link.click()
    race_table = driver.find_element(By.TAG_NAME, "tbody")
    race_rows = race_table.find_elements(By.TAG_NAME, "tr")
    for race_row in race_rows:
        content = race_row.find_elements(By.TAG_NAME, "td")
        athlete_name = str(content[1].find_element(By.TAG_NAME, "a").text)
        athlete_time = str(content[-1].text)
        print(marathon_name, date, athlete_name, athlete_time)
        data.append([marathon_name, date, athlete_name, athlete_time])
    driver.back()
    sleep(1)

# athlete_table = driver.find_element(By.TAG_NAME, "tbody")
# rows = athlete_table.find_elements(By.TAG_NAME, "tr")
# for row in rows:
#     content = row.find_elements(By.TAG_NAME, "td")
#     athlete_name = content[1].find_element(By.TAG_NAME, "a").text
#     athlete_time = content[-1].text
#     print(athlete_name, athlete_time)
```
